TAGS/REGEX

Remove (modern scribal notations): ! (certain reading) ? (questionable reading) / (line divider) : OR . (word divider) < > (scribal insertions, but keep the text in translit / translations) ˹ ˺ (partially broken signs, to be removed from transliteration) [ ] (remove from document level transliteration. e.g. [KÙ.BABBAR] → KÙ.BABBAR)

```python
import re
def clean_scribal_notations(text):
    if pd.isna(text):
        return text

    text = re.sub(r'<([^>]*)>', r'\1', text)
    text = re.sub(r'\[([^\]]*)\]', r'\1', text)
    text = re.sub(r'[!?/:.\u02F9\u02FA]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()

    return text
```

```python
import re
import pandas as pd

SUBSCRIPT_MAP = str.maketrans(
    "₀₁₂₃₄₅₆₇₈₉",
    "0123456789"
)

def normalize_transliteration(text):
    if pd.isna(text):
        return text

    # 1. Explicit gaps
    text = re.sub(r'\[x\]', '<gap>', text)

    # 2. Big gaps
    text = re.sub(r'\[\s*…\s*…\s*\]', '<big_gap>', text)
    text = re.sub(r'…', '<big_gap>', text)

    # 3. Remove gloss braces
    text = re.sub(r'\b(\w+)\s*\{\1\}', r'\1', text)

    
    # 4. Normalize subscripts
    text = text.translate(SUBSCRIPT_MAP)

    # 5. Remove broken-sign markers
    text = re.sub(r'[\u02F9\u02FA]', '', text)

    # 6. Normalize modifier apostrophe
    text = re.sub(r'\u02BE', "'", text)

    # 7. Remove editorial symbols
    text = re.sub(r'[!?/:.]', '', text)

    # 8. Normalize whitespace
    text = re.sub(r'\s+', ' ', text).strip()

    return text
```

```python
import re
import pandas as pd

def join_hyphenated_symbols(text):
    if pd.isna(text):
        return text

    # Join tokens split by hyphens
    # ma-nu-ba-lúm-a-šur → manubalúmašur
    # ṣí-lá-(d)IM → ṣílá(d)IM
    text = re.sub(r'(?<=\S)-(?=\S)', '', text)

    return text
```

```python
import re
import pandas as pd

ATF_ASCII_TO_UNICODE = {
    r'sz': 'š',
    r'SZ': 'Š',

    r's,': 'ṣ',
    r'S,': 'Ṣ',

    r't,': 'ṭ',
    r'T,': 'Ṭ',

    r"s'": 'ś',
    r"S'": 'Ś',

    r'h,': 'ḫ',
    r'H,': 'Ḫ',

    r'j': 'ŋ',
    r'J': 'Ŋ',

    r'\u02BE': 'ʾ',  # modifier apostrophe (if present)
}

def normalize_atf_characters(text):
    if pd.isna(text):
        return text

    for ascii_pat, uni_char in ATF_ASCII_TO_UNICODE.items():
        text = re.sub(ascii_pat, uni_char, text)

    return text
```

```python
import pandas as pd

def normalize_subscript_x(text):
    if pd.isna(text):
        return text

    # Unicode subscript x (U+208A) → uppercase X
    text = text.replace('\u208A', 'X')

    return text
```

```python
import re
import pandas as pd

def remove_sign_name_markers(text):
    if pd.isna(text):
        return text

    # Remove $ or ~ only when they precede an uppercase sign name
    text = re.sub(r'([$~])(?=[A-Z])', '', text)

    return text
```

```python
import re
import pandas as pd

def normalize_qualified_x_signs(text):
    if pd.isna(text):
        return text

    # Join qualified signs: ba_x(PI) → ba_xPI , ZAX(LAK384) → ZAXLAK384
    text = re.sub(
        r'([A-Za-z0-9ₓx]+)\(([^)]+)\)',
        r'\1\2',
        text
    )

    return text
```

```python
from nltk.corpus import stopwords

def remove_stopwords(text):
    new_text = []

    for word in text.split():
        if word in stopwords.words('english'):
            new_text.append('')
        else:
            new_text.append(word)
    
    x = new_text[:]
    new_text.clear()
    return " ".join(x)
```
