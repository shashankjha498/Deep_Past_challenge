```python
import pandas as pd
import numpy as np

def add_para_to_sentence(df: pd.DataFrame) -> pd.DataFrame:
    """
    Adds 'para_to_sentence' column by splitting transliteration paragraphs into sentences.

    Assumes:
    - Rows with same text_uuid share the same transliteration paragraph.
    - first_word_number and first_word_obj_in_text are 1-based word indices (int or NaN).
    - Multiple rows per paragraph define all sentence starts collectively.
    """
    def process_group(group: pd.DataFrame) -> pd.DataFrame:
        if len(group) == 0:
            return group

        para_text = group['transliteration'].iloc[0]
        tokens = para_text.split()
        n_tokens = len(tokens)

        
        valid_starts_1based = []
        for _, row in group.iterrows():
            start_num = int(row['first_word_number'])
            obj_num = row['first_word_obj_in_text']
            if pd.isna(obj_num) or int(obj_num) != start_num:
                valid_starts_1based.append(start_num)

        valid_starts_1based = sorted(set([s for s in valid_starts_1based if s <= n_tokens]))

        
        all_sentences = []
        for i, start in enumerate(valid_starts_1based):
            end = valid_starts_1based[i + 1] if i + 1 < len(valid_starts_1based) else n_tokens + 1
            sent_tokens = tokens[start - 1 : end - 1]
            all_sentences.append(' '.join(sent_tokens))

        
        result = group.copy()
        for idx, row in result.iterrows():
            start_num = int(row['first_word_number'])
            for s_idx, s_start in enumerate(valid_starts_1based):
                if s_start == start_num:
                    result.at[idx, 'para_to_sentence'] = all_sentences[s_idx]
                    break
            else:
                result.at[idx, 'para_to_sentence'] = ''  

        return result

    
    return df.groupby('text_uuid', group_keys=False).apply(process_group, include_groups=False).reset_index(drop=True)


result_df = add_para_to_sentence(final_df)
```

```python
import pandas as pd
from sentence_transformers import SentenceTransformer, util
import numpy as np


def validate_sentence_reconstruction(df: pd.DataFrame) -> pd.DataFrame:
    """
    For each oare_id:
    1. Concat all translation_sent → combined_sent
    2. Compare combined_sent vs translation via cosine similarity
    """
  
    if df['translation_sent'].dtype == 'object':
        df['translation_sent_flat'] = df['translation_sent'].apply(
            lambda x: ' '.join(x) if isinstance(x, list) else str(x)
        )
    else:
        df['translation_sent_flat'] = df['translation_sent'].astype(str)

    
    grouped = df.groupby('oare_id').agg({
        'translation_sent_flat': lambda x: ' '.join(x.dropna().unique()),  
        'translation': 'first'  
    }).reset_index()

    
    model = SentenceTransformer('all-MiniLM-L6-v2')
    combined_sents = grouped['translation_sent_flat'].tolist()
    para_trans = grouped['translation'].dropna().tolist()

    if combined_sents:
        embeds1 = model.encode(combined_sents, show_progress_bar=True)
        embeds2 = model.encode(para_trans, show_progress_bar=True)
        similarities = util.pytorch_cos_sim(embeds1, embeds2).diag().cpu().numpy()
        grouped['similarity_score'] = similarities
    else:
        grouped['similarity_score'] = np.nan

    return grouped[['oare_id', 'similarity_score', 'translation_sent_flat', 'translation']]


result_df_2 = validate_sentence_reconstruction(result_df)
print(result_df_2[['oare_id', 'similarity_score']].round(3))
```

```python
def set_translation_final(df):
    """ONLY creates translation_final column per your spec."""

    df_work = df.copy()
    df_work['oare_id'] = df_work['oare_id'].astype(str)

    def process_group(group):
        if len(group) == 1:
            
            group['translation_final'] = group['translation'].iloc[0]
        else:
            
            group['translation_final'] = group['translation_sent']

            
            if len(group) > 2 and 'similarity_score' in group.columns:
                sim_mean = pd.to_numeric(group['similarity_score'], errors='coerce').mean()
                if not pd.isna(sim_mean) and sim_mean < 0.9:
                    print(f"Low similarity oare_id {group.name}: {sim_mean:.3f}")

        return group

    result = df_work.groupby('oare_id').apply(process_group).reset_index(drop=True)

    print(f"Added translation_final to {len(result)} rows")
    return result


df_final = set_translation_final(result_df_merged)
df_final
```
